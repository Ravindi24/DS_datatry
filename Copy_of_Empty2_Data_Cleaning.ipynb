**Dealing with Outliers**

```python
train_path="/content/drive/MyDrive/Dataset_try/iris_Outliers.csv"
df=pd.read_csv(train_path)
df.head()
```

Identify outliers-using Boxplot (IQR method)

```python
import matplotlib.pyplot as plt
```

```python
df.iloc[:,:4].boxplot()
plt.show()
```

```python
df.drop("Species",axis=1,inplace=True)
df
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
LL=Q1-(1.5*IQR)
UL=Q3+(1.5*IQR)
```

```python
LL
```

```python
UL
```

```python
dfn=df
dfn.head()
```

```python
dfr=(dfn<LL) | (dfn>UL)  # OR operator
dfr.head(15)
```

```python
out_row=((dfn<LL) | (dfn>UL)).any(axis=1) # rows having at least one outlier
```

```python
# taking rows without having outliers
df_out_free=df[~out_row]  # ~=NOT
df_out_free   # dataset by removing outliers
```
