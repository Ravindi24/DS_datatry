***Movie Recomendation system using Colaborative Filtering***

```python
import pandas as pd
import numpy as np
```

**Get Data**

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('/content/drive/MyDrive/ML_data3/u.data',sep='\t',names=column_names)
df.head()
```

```python
movie_titles = pd.read_csv("/content/drive/MyDrive/ML_data3/Movie_Id_Titles.txt",delimiter=',')
movie_titles.head()
```

```python
df =pd.merge(df,movie_titles,on='item_id')
df.head()
```

**Recommending Similar Movies**

```python
moviemat =df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
#most rated movie:
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
# Now let's grab the user ratings for those two movies:

starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

 corrwith() method to get correlations between two pandas series:

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

Now if we sort the dataframe by correlation, we should get the most similar movies, however note that we get some results that don't really make sense. This is because there are a lot of movies only watched once by users who also watched star wars (it was the most popular movie).

```python
corr_starwars.sort_values('Correlation',ascending=False).head(15)
```

Let's fix this by filtering out movies that have less than 100 reviews (this value was chosen based off the histogram from earlier).

```python
corr_starwars= corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
# sort values
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

*** *For those who watch the 'Star Wars(1977)', the next movie recommended by the system is 'Empire Strikes Back, The (1980)' * ***

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

***For those who watch the 'Liar Liar (1997)', the next movie recommended by the system is 'Batman Forever (1995)	' ***
